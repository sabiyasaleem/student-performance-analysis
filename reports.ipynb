# Student Performance Data Cleaning & Analysis Report

## Objective
##### The goal of this project is to clean, preprocess, and analyze a student performance dataset to make it suitable for analysis and derive meaningful insights about factors affecting academic performance.

## Dataset Description
##### The dataset contains information about students, including:

##### 1. Demographics (age, gender, address)
##### 2. Family background (parents’ education, jobs)
##### 3. Academic data (G1, G2, G3 grades)
##### 4. Lifestyle factors (study time, alcohol consumption, social outings)

### Key Target Variable:
##### Final Grade (G3 / final_grade) → Represents overall student performance

# Data Cleaning Steps

### Data Loading
##### Dataset loaded using Pandas
##### Separator (;) handled correctly

### Column Standardization
##### Renamed columns to meaningful names
##### (e.g., G3 → final_grade)

### Text Cleaning
##### Removed extra spaces and quotes
##### Converted text to lowercase for consistency

### Numeric Conversion
##### Converted relevant columns to numeric format
##### Invalid values handled using errors='coerce'

### Missing Values
##### Identified using .isnull()
##### Removed using dropna()

### Duplicate Removal
##### Duplicate rows removed to ensure data integrity

### Value Standardization
#### Encoded values like:
##### u → urban
##### r → rural
##### gp → gabriel pereir

# Exploratory Data Analysis (EDA)

### Study Time vs Final Grade

##### Weak positive correlation observed
##### Students who study more tend to perform slightly better

### However, performance is influenced by multiple factors

##### Grade Progression (G1 → G2 → G3)

##### Strong relationship between grades across periods
##### Early performance impacts final outcome

### Alcohol Consumption Impact

##### Higher alcohol consumption shows a tendency toward lower grades

### Family Influence

##### Parental education has a moderate impact on student performance

# Key Insights

## Early Performance is the Strongest Predictor
##### Students with high G1 and G2 almost always have high G3
##### Very few students drastically improve or drop at the end

### Insight:
##### Early academic performance is a strong indicator of final outcomes.

## Study Time Has Diminishing Returns
##### Moving from low → medium study time improves grades
##### But high to very high study time doesn’t always increase performance

### Insight:
##### Beyond a certain point, increasing study time has limited impact.

### Alcohol Consumption Negatively Impacts Grades
##### Higher weekday (Dalc) and weekend (Walc) alcohol use → generally lower final grades

### Insight:
##### Lifestyle habits play a measurable role in academic success.

### Parental Education Influence
##### Students with more educated parents tend to perform slightly better

### Insight:
##### Family educational background contributes to academic support and outcomes.

### Urban vs Rural Students

##### Urban students may show slightly better performance (observed in data)

### Insight:
##### Access to resources (internet, schools, support) may influence performance.

### Absences Strongly Affect Grades

##### Higher absences → lower grades

### Insight:
##### Attendance is a critical factor in academic success.

### Social Life Trade-off

##### More social outings  often correlates with lower grades

### Insight:
##### Excessive social activity may reduce study time and performance.

### Health Has Moderate Impact

##### Students with better health tend to perform slightly better

### Insight:
##### Physical well-being supports academic performance.

## No Single Factor Dominates

##### No variable alone perfectly predicts grades

### Insight:
##### Student performance is multi-dimensional (study + lifestyle + family)

### The analysis reveals that academic performance is not driven by a single factor but by a combination of study habits, prior performance, lifestyle choices, and family background. Among these, prior academic results (G1 and G2) are the most significant predictors of final outcomes.


# Limitations of the Project
### Small Dataset Size
##### The dataset is limited, so results may not represent all students.
### Limited Variables
##### Important factors like mental health, teaching quality, and income are not included.
### Region-Specific Data
##### Data comes from a specific area, so findings may not apply everywhere.
### Correlation-Based Analysis
##### The project shows relationships but cannot prove cause and effect.
### Static (Historical) Data
##### The dataset reflects past student data and may not represent current trends or behaviors

# Conclusion
##### This project successfully cleaned and preapered the student dataset for analysis.
##### The findings highlight that student performance is influenced by a combination of academic habits,lifestyle choices, and family background.
